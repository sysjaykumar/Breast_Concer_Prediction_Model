Importing Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

Data Collection & Preprocessing

```python
#Loading data from sklearn\
breast_cancer = sklearn.datasets.load_breast_cancer()
print(breast_cancer)
```

```python
#loading data to dataframe
data_frame = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
data_frame.head()
```

```python
# adding the target columns to be data frame
data_frame['label']=breast_cancer.target

#print last 5 rows of dataframe
data_frame.tail()
```

```python
#number of rows & column in dataframe
data_frame.shape
```

```python
# getting another information
data_frame.info()
```

```python
# Checking for missing value
data_frame.isnull().sum()
```

```python
# some statstical measures
data_frame.describe()
```

```python
#Checking distribution of Target values
data_frame['label'].value_counts()
```

1 --> Begin
0 --> maligant

```python
data_frame.groupby('label').mean()
```

Seerating the Features and target value

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

Spliting the data into training and testing data

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y , test_size=0.2)
```

```python
print(X.shape, x_train.shape, x_test.shape)
```

Standerization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x_train_sd = scaler.fit_transform(x_train)
x_test_sd = scaler.fit_transform(x_test)
```

```python
print(x_train_sd.shape, x_test_sd.shape)
```

Building the Neural Network

```python
#importing tensorflow and Keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
from keras.layers import Input, Flatten, Dense
```

```python
# setting up the layers of Neural network
model = keras.Sequential([
    Input(shape=(30,)),            
    Flatten(),
    Dense(20, activation='relu'),
    Dense(10, activation='sigmoid'),
    Dense(2, activation='sigmoid')
])
```

```python
#Compile the Neural Network
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
#Training the Neural Network
history = model.fit(x_train_sd, y_train, validation_split=0.1, epochs=10)
```

Visualizing accuracy or loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Training Accuracy','Validation Accuracy'],loc='upper right')
plt.grid(True)
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Training Accuracy','Validation Accuracy'],loc='upper right')
plt.grid(True)
plt.show()
```

Accuracy of the mode of this data

```python
loss, acccuracy = model.evaluate(x_test_sd, y_test)
print(acccuracy)
print(x_test_sd.shape)
print(x_test_sd[0])
```

```python
y_pred = model.predict(x_test_sd)
print(y_pred[0])
```

```python
print(y_pred.shape)
print(y_pred)
```

model.predict() given the prediction probability to each class for that data point

```python
#converting the predictionprobability tdo class label
Y_pred_label = [np.argmax(i) for i in y_pred]
print(Y_pred_label)
```
